```python
diagonal = CSR_mat_full.diagonal().tolist()
sortedlist = sorted(range(len(diagonal)), key=diagonal.__getitem__, reverse=True)
```

```python
%matplotlib notebook
from ipywidgets import *
fig = matplotlib.pyplot.figure()
ax = fig.add_subplot(111)
matplotlib.pyplot.ion()
def interactplot(i):
    ax.clear()
    ax.plot(diagonal[sortedlist[i]-200:sortedlist[i]+200])
    ax.plot(CSR_mat_full[sortedlist[i]-5:sortedlist[i]+5,].toarray().sum(axis=0).tolist()[sortedlist[i]-200:sortedlist[i]+200])
    
    fig.canvas.draw()
    
interact(interactplot,i=widgets.IntSlider(min=0,max=500,step=1,value=1));

#matplotlib.pyplot.plot(CSR_mat_full[index_max-5:index_max+5,].toarray().sum(axis=0).tolist()[index_max-200:index_max+200])
```

```python
%matplotlib qt
#scatter of lenght vs number of reads
fig, ax = matplotlib.pyplot.subplots()
lengths=[]
for i in frag_prop:
    lengths.append(i[3])
ax.scatter(lengths[:584662],CSR_mat_full.diagonal().tolist()[:584662],alpha=0.01)
ax.set_xlim((0, 5000))
ax.set_ylim((0, 5000))
fig.show()
```

```python
#preparing data for distribution plots
distribution = [[] for i in range(10000)]
diagonal=CSR_mat_full.diagonal().tolist()
lengths=[]
for i in frag_prop:
    lengths.append(i[3])
       
for i in range(len(lengths)):
    try:
        distribution[lengths[i]].append(diagonal[i])
    except:
        continue

results = []        
for i in distribution:
    if i:
        results.append(statistics.mean(i))
    else:
        results.append(1000)
        
for i in range(len(lengths)):
    try:
        if diagonal[i] != 0:
            distribution[lengths[i]].append(diagonal[i])
    except:
        continue        

results2 = []        
for i in distribution:
    if i:
        results2.append(statistics.mean(i))
    else:
        results2.append(1000)        
```

```python
#plotting avarage reads vs length
%matplotlib qt
fig, ax = matplotlib.pyplot.subplots()
ax.scatter(range(10000),results)
ax.scatter(range(10000),results2)
fig.show()
```

```python
#fragment length distribution
%matplotlib qt
fig, ax = matplotlib.pyplot.subplots()
ax.hist(lengths, bins= 1000,range=(0, 10000))

fig.show()
```

just a section to make a nicer plot for the bamfingerprints

```python
with open("./testdata/fingerprints_nodup.tab", "r") as fingerprints:
    #skip two lines
    next(fingerprints)
    next(fingerprints)
    samples=["NaiveT_27ac_B2_T1","NaiveT_27ac_B3_T1","lgs000379_H3K27ac","Jurkat_27ac_Arima","HaCat_27ac_st_10","HaCat_27ac_unst_10","MyLa_27ac_10_lane1","MyLa_27ac_10_both","lgs101645_H3K27ac"]
    values={}
    for i in range(len(samples)):
        values[samples[i]]=[]
    for line in fingerprints:
        bins=line.split()
        for i in range(len(samples)):
            values[samples[i]].append(int(bins[i].strip()))
    
    
```

```python
for i in range(len(values)):
    values[samples[i]]=sorted(values[samples[i]])
values
```

```python
%matplotlib qt
for key in values:
    matplotlib.pyplot.plot(numpy.array(values[key]).cumsum()/numpy.array(values[key]).sum(),label=key)
    
matplotlib.pyplot.legend()
```
